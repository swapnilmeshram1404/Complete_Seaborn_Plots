### DISTRIBUTION PLOT

Distribution plot is used to display a ssingle continuous feature 
and help visualize properties such as deviation and average value.

There are 3 main distribution plot types:
    
    1) Rug Plot
    2) Histogranm
    3) KDE Plot

```python
np.random.seed(42)

sample_ages = np.random.randint(0,100,200)

sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages , columns = ["age"])
```

```python
sample_ages
```

```python
plt.figure(dpi = 100)

sns.rugplot(data = sample_ages , x = "age", height = 0.5, color = "black" )
```

```python
sns.displot(data = sample_ages , bins = 30,
            lw = 2 , edgecolor = "black" , kde = True)
```

```python
sns.kdeplot(data = sample_ages , x = "age") 

# here ages are shown -ve and above 100 due to gausian distribution 
```

```python
sns.kdeplot(data = sample_ages , x = "age" , clip = [0,100] , shade = True) 

#here, -ve ages and ages above 100 are cliped
```
